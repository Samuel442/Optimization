```python
import numpy as np
from scipy.interpolate import CubicSpline

# As bibliotecas numpy e scipy.interpolate são importadas. A biblioteca numpy 
# é amplamente utilizada para trabalhar com 
# arrays e operações numéricas, enquanto scipy.interpolate contém funções para interpolação.

# Pontos de controle da spline cúbica
x = np.array([0, 1, 2, 3, 7])
y = np.array([0, 2, 1, 2, 0])

# São definidos dois arrays x e y que representam os pontos de controle 
# da spline cúbica. Neste caso, temos cinco pontos: (0,0), (1,3), (2,1), (3,2) e (4,0).

# Criação da spline cúbica
#A função CubicSpline é utilizada para criar uma instância de spline cúbica a
#partir dos pontos de controle fornecidos. Essa função realiza o cálculo dos 
#coeficientes da spline cúbica que melhor se ajustam aos pontos de controle.
spline = CubicSpline(x, y)

# Pontos para avaliar a spline cúbica
#Aqui, um array x_eval é criado usando a função linspace do NumPy, que gera
#200 valores igualmente espaçados entre 0 e 4. Esses valores serão usados para 
#avaliar a spline cúbica. A função spline(x_eval) é chamada para calcular os 
#valores de y_eval, que são as coordenadas y correspondentes aos pontos x_eval
#na spline cúbica.
x_eval = np.linspace(0, 4, num=200)
y_eval = spline(x_eval)

# Plot da spline cúbica
#As bibliotecas matplotlib.pyplot são importadas para plotar os gráficos. 
#Em seguida, são utilizadas as funções plot para plotar os pontos de controle 
#(x e y) como pontos no gráfico ('o' significa marcadores circulares) e a spline 
#cúbica (x_eval e y_eval) como uma linha contínua. A função legend é usada para 
#adicionar uma legenda ao gráfico e show é chamada para exibir o gráfico resultante.
import matplotlib.pyplot as plt
plt.plot(x, y, 'o', label='Pontos de controle')
plt.plot(x_eval, y_eval, label='Spline cúbica')
plt.legend()
plt.show()
```
